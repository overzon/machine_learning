```python
from sklearn import datasets
iris = datasets.load_iris()
#print(iris.DESCR)
iris
```

```python
import numpy as np
from sklearn.linear_model import LogisticRegression
x = iris.data[:,3].reshape(-1,1) #petal width
y = (iris.target == 2).astype(np.int)#tran t or f to 0 or 1
log_re = LogisticRegression(random_state=42)
log_re.fit(x,y)
```

```python
xnew = np.linspace(0,3,1000).reshape(-1,1)
phat = log_re.predict_proba(xnew)
phat
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(xnew,phat[:,1], label = "virginica")
plt.plot(xnew,phat[:,0], label = "Not virginica")
```

```python
import numpy as np
from sklearn.linear_model import LogisticRegression
```

```python
x = iris.data[:,2:]
y = (iris.target == 2).astype(np.int)#tran t or f to 0 or 1
log_re = LogisticRegression(random_state=42)
log_re.fit(x,y)
xnew = np.c_[np.linspace(0,3,1000),np.linspace(0,3,1000)]
phat = log_re.predict(xnew)
plt.plot(x[y==0,0],x[y==0,1], "bs" , label="Not Virginica")
plt.plot(x[y==1,0],x[y==1,1], "rs" , label="Virginica")
plt.plot(xnew[phat==0,0],xnew[phat==0,1], "bs" , label="Not Virginica")
plt.plot(xnew[phat==1,0],xnew[phat==1,1], "rs" , label="Virginica")
```
