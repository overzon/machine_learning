```python
from sklearn import datasets
iris = datasets.load_iris()
#print(iris.DESCR)
iris
```

```python
import numpy as np
from sklearn.linear_model import LogisticRegression
x = iris.data[:,3].reshape(-1,1) #petal width
y = (iris.target == 2).astype(np.int)#tran t or f to 0 or 1
log_re = LogisticRegression(random_state=42)
log_re.fit(x,y)
```

```python
xnew = np.linspace(0,3,1000).reshape(-1,1)
phat = log_re.predict_proba(xnew)
phat
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(xnew,phat[:,1], label = "virginica")
plt.plot(xnew,phat[:,0], label = "Not virginica")
```

```python
from sklearn import datasets
import matplotlib.pyplot as plt
import numpy as np
from sklearn.svm import SVC
iris = datasets.load_iris()
x = iris.data[:,2:].reshape(-1,2) #petal width
y = (iris.target == 2).astype(np.int)#tran t or f to 0 or 1
classifier = SVC(kernel="linear",C=100000)
classifier.fit(x,y)

w = classifier.coef_[0]
b = classifier.intercept_[0]
# xnew = np.linspace(0,3,1000).reshape(-1,1)
# phat = log_re.predict_proba(xnew)
x0 = np.linspace(0,8,200)
boundary = -w[0]/w[1] * x0 - b/w[1]

# wTx + b = 0
# w0x0 + w1x1 + b = 0
# x1 = f(x0) = (-w0x0 - b)/w1

plt.plot(x0, boundary)
plt.plot(x[y==0,0],x[y==0,1], "bs" , label="Not Virginica")
plt.plot(x[y==1,0],x[y==1,1], "rs" , label="Virginica")
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

x,y = datasets.make_moons(n_samples=100, noise=0.15, random_state=42)
plt.plot(x[y==0,0],x[y==0,1], "bs" , label="blude")
plt.plot(x[y==1,0],x[y==1,1], "rs" , label="red")

classifier = Pipeline([
    ("scaler", StandardScaler()),
    ("svm",SVC(kernel="poly",degree=3,C=5,coef0=1))
])
classifier.fit(x,y)

x0s = np.linspace(-1.5,2.5,100)
x1s = np.linspace(-1, 1.5,100)
x0, x1 = np.meshgrid(x0s,x1s)
xpred = np.c_[x0.ravel(), x1.ravel()]
ypred = classifier.predict(xpred).reshape(x0.shape)
ydecision = classifier.decision_function(xpred).reshape(x0.shape)
plt.contourf(x0,x1, ydecision, cmap=plt.cm.brg,alpha=0.2)
plt.contourf(x0,x1,ypred,cmap=plt.cm.brg,alpha=0.2)
```
